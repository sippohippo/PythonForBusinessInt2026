## Importing needed libraries, loading and cleaning the data

```python
# Importing pandas and what we need from scikit-learn

import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
# Loading the data

df = pd.read_csv("https://github.com/allisonhorst/palmerpenguins/raw/5b5891f01b52ae26ad8cb9755ec93672f49328a8/data/penguins_size.csv")
```

```python
# Cleaning the data quickly by removing rows with missing values

# You will learn how to do this properly during exercise 2

df = df.dropna()
```

```python
# This is what the data looks like

df.head()
```

## Splitting the dataset and training the model

```python
# Taking the numeric values into one dataframe and labels (species) into another

X = df[["culmen_length_mm", "culmen_depth_mm", "flipper_length_mm", "body_mass_g"]]   # the features
y = df["species_short"] # the labels
```

```python
# Let's split the data into a training and test set. We will use 70% for training and 30% for testing.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# Choosing the model

model = RandomForestClassifier(random_state = 42)
```

```python
# Training the model

model.fit(X_train, y_train)
```

## Evaluating how well the model works with the test set

```python
# Evaluating performance

pred = model.predict(X_test)
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```

You might be wondering what precision, recall and f1-score means. They are all different performance metrics that tell how well the ML model is classifying the penguins. We will go through these and much more during the next lecture. For now it is enough to know that such scores are suspiciously high and even though it looks like the model is working well, it might not be so good in identifying the different species when it would be shown real new data from outside of the Palmer Penguins dataset.
