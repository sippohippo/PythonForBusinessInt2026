## 0. Getting Started with Jupyter Notebooks

Before we go into Python, let's quickly go through the most important features of **Jupyter Notebook**.  
A notebook is made up of **cells**. There are two main types:

- **Code cells** — where you write and run Python code (they have `In [ ]:` on the left).
- **Markdown cells** — where you write formatted text, like the explanations you're reading now.

### Essential keyboard shortcuts

Jupyter has two modes:
- **Edit mode** — you're typing inside a cell. For code cells click once, for markdown cells double click. Press `Esc` to leave. For Markdown cells to render again after edit mode, press shift `Shift + Enter`
- **Command mode** — you're navigating between cells. Press `Enter` to edit a cell.

| Shortcut | Mode | What it does |
|----------|------|--------------|
| `Shift + Enter` | Either | **Run** the current cell and move to the next one |
| `Ctrl + Enter` | Either | **Run** the current cell and stay on it |
| `Alt + Enter` | Either | **Run** the current cell and insert a new cell below |
| `A` | Command | Insert a new cell **above** |
| `B` | Command | Insert a new cell **below** |
| `DD` (press D twice) | Command | **Delete** the current cell |
| `M` | Command | Convert cell to **Markdown** |
| `Y` | Command | Convert cell to **Code** |
| `Z` | Command | **Undo** cell deletion |
| `Ctrl + S` | Either | **Save** the notebook |

Note, on a Mac use `Command` instead of `Ctrl`

### Useful tips

- **Run cells in order.** Code cells often depend on variables created in earlier cells. If you get a `NameError`, try running the cells above first.
- **Restart the kernel** if things get messy. Go to **Kernel → Restart & Clear Output** to start fresh. This erases all variables from memory but keeps your code.
- **Kernel → Restart & Run All** re-runs every cell from top to bottom — handy for a clean check of your entire notebook.
- **Tab completion:** While typing in a code cell, press `Tab` to see suggestions (e.g., type `pri` then `Tab` to see `print`).
- **Help on any function:** Type a function name followed by `?` and run the cell (e.g., `print?`) to see its documentation. You can also use `Shift + Tab` while your cursor is on a function name.
- **Add comments** in code cells by starting a line with `#`. Comments are ignored by Python and are just notes for humans.

## 1. Variables and Data Types

In programming, a **variable** is a location for storing a value, such as a string or a number. This value can then be used later in the code, by referring to the variable. The value within the variable can also be changed later.

Python has several basic **data types**:

| Type | What it stores | Example |
|------|----------------|---------|
| `int` | Whole numbers | `42` |
| `float` | Decimal numbers | `3.14` |
| `str` | Text (strings) | `"hello"` |
| `bool` | True / False | `True` |

```python
# To create a variables just use the = sign
company_name = "Acme Corp"    # str  (text — always in quotes)
revenue = 5000000              # int  (whole number)
growth_rate = 0.12             # float (decimal number)
is_profitable = True           # bool (True or False)
 
# Print them out
print(company_name)
print(revenue)
print(growth_rate)
print(is_profitable)
```

```python
# You can check a variable's type with type()
print(type(company_name))   # <class 'str'>
print(type(revenue))        # <class 'int'>
print(type(growth_rate))    # <class 'float'>
print(type(is_profitable))  # <class 'bool'>
```

### Naming rules & conventions

- Variable names **must** start with a letter or underscore, not a number.
- They **cannot** contain spaces — use underscores instead: `total_sales`, not `total sales`.
- Python is **case-sensitive**: `Revenue` and `revenue` are different variables.
- Use descriptive names: `q3_revenue` is much better than `x`.

## 2. Expressions and Statements

- An **expression** is any piece of code that *produces* a value:  
  `3 + 4`, `revenue * growth_rate`, `len("hello")`

- A **statement** is a command that has some effect. Examples include calling a function (when not part of an expression) or assigning values:  
  `x = 10` (assigns a value), `print(x)` (displays output)


```python
# Arithmetic expressions
print(10 + 3)     # Addition
print(10 - 3)     # Subtraction
print(10 * 3)     # Multiplication
print(10 / 3)     # Division (always returns a float)
print(10 ** 2)    # Exponentiation (10 squared)
print(10 % 3)     # Modulo — the remainder after division
print(10 // 3)    # Floor division — division rounded down
```

```python
# String statements
first = "Mikko"
last = "Mallikas"
full_name = first + " " + last   # Concatenation (joining strings)

print(full_name)

# Note, f-strings (demonstrated below) are a convenient way to embed variables inside text
greeting = f"Hello, {full_name}! Your name has {len(full_name)} characters."
print(greeting)
```

```python
# Comparison expressions — they produce True or False
print(10 > 5)      # True
print(10 == 10)    # True  (== checks equality; = assigns)
print(10 != 5)     # True  (not equal)
print(10 < 5)      # False
```

```python
# Statements can store the result of an expression
projected_revenue = revenue * (1 + growth_rate)
print(f"Projected revenue: ${projected_revenue:,.0f}")
```

## 3. Data Structures: Lists, Dictionaries, and Tuples

So far, each variable has held *one* value. Data structures let you store **collections** of values in a single variable.

| Structure | Syntax | Ordered? | Mutable? |
|-----------|--------|----------|----------|
| **List** | `[ ]` | Yes | Yes (can change) |
| **Tuple** | `( )` | Yes | No (fixed) |
| **Dictionary** | `{ }` | Yes* | Yes |

*Dictionaries preserve insertion order in Python 3.7+.

### 3a. Lists

A list is an **ordered, changeable** collection. You can add, remove, or modify items after creation.

```python
# Creating a list
products = ["Laptop", "Phone", "Earphones"]
print(products)

# Accessing items by index (Python counts from 0!)
print(products[0])    # First item: "Laptop"
print(products[-1])   # Last item:  "Earphones"
```

```python
# Modifying a list
print("\nAfter adding Charger")
products.append("Charger")    # Add to the end
print(products)

print("\nAfter changing item")
products[1] = "PS5"      # Change an item
print(products)

products.remove("Laptop")      # Remove by value
print("\nAfter removing laptop")
print(products)

print(f"Number of products: {len(products)}")
```

```python
# Lists can hold mixed types (though usually you keep them consistent)
mixed = ["Acme", 2024, True, 3.14]
print(mixed)
```

### 3b. Dictionaries

A dictionary stores data as **key-value pairs** — like a real dictionary where each word (key) has a definition (value).

```python
# Creating a dictionary
customer = {
    "name": "Alice Johnson",
    "email": "alice@example.com",
    "age": 34,
    "is_premium": True
}

# Accessing values by key
print(customer["name"])
print(customer["email"])
```

```python
# Adding and updating entries
customer["phone"] = "555-0123"       # Add new key
customer["age"] = 35                  # Update existing key
print(customer)

# Useful methods
print(customer.keys())    # All keys
print(customer.values())  # All values
```

### 3c. Tuples

A tuple looks like a list but uses parentheses `( )` and **cannot be changed** after creation.  
Use tuples when you want to make sure data stays fixed — for example, the coordinates of a store location.

```python
# Creating a tuple
store_location = (40.7128, -74.0060)   # (latitude, longitude) for New York
print(store_location)
print(store_location[0])               # Access by index, just like a list

# Trying to change a tuple will cause an error — uncomment the line below to see:
# store_location[0] = 41.0   # ← TypeError!
```

## 4. Conditional Statements

When programming, we often want code to behave one way if some conditions are met, and another way if not.
Python uses `if`, `elif` (else-if), and `else` to execute different code depending on whether a condition is `True` or `False`.

**Key syntax rule:** the code block inside an `if` must be **indented** (typically 4 spaces). Python uses indentation instead of braces `{ }` to define blocks.

```python
temperature = 22

if temperature > 25:
    print("It's hot outside!")
elif temperature > 15:
    print("Nice weather!")
else:
    print("Better bring a jacket!")
```

```python
# A business example: simple credit decision
credit_score = 720
annual_income = 85000

if credit_score >= 750 and annual_income >= 50000:
    decision = "Approved - Premium"
elif credit_score >= 650 and annual_income >= 40000:
    decision = "Approved - Standard"
else:
    decision = "Denied"

print(f"Credit decision: {decision}")
```

Notice the keywords `and` / `or` for combining conditions:
- `and` — **both** conditions must be True
- `or` — **at least one** condition must be True
- `not` — flips True to False and vice versa

## 5. Loops: `for` and `while`

Loops let you **repeat** a block of code multiple times — so you don't have to write the same thing over and over.

### 5a. `for` loops

A `for` loop iterates over each item in a sequence (like a list, a string, or a range of numbers).

```python
# Looping through a list
departments = ["Marketing", "Finance", "Operations", "Management"]

for dept in departments:
    print(f"Welcome to the department of {dept}!")
```

```python
# Looping through a range of numbers
# range(5) generates: 0, 1, 2, 3, 4
for i in range(5):
    print(f"Iteration {i}")
```

```python
# A practical example: calculate total revenue from a list of sales
monthly_sales = [12000, 15000, 9000, 18000, 21000, 14000]

total = 0
for sale in monthly_sales:
    total = total + sale   # (can also write: total += sale)

print(f"Total revenue: {total:}€")
print(f"Average monthly: {total / len(monthly_sales):.0f}€")
```

Note though that lists and other data structures often have built in functions for e.g. calculating the sum of all elements.

```python
# This calculates the average without using a loop:
print(f"Total revenue: {sum(monthly_sales):}€")
print(f"Average monthly: {sum(monthly_sales) / len(monthly_sales):.0f}€")
```

### 5b. `while` loops

A `while` loop keeps running **as long as** a condition is `True`.  
Be careful: if the condition never becomes `False`, the loop will run forever!

```python
# Simulating compound interest growth
balance = 10000    # starting investment
target = 20000     # goal
rate = 0.07        # 7% annual return
years = 0

while balance < target:
    balance = balance * (1 + rate)
    years += 1
    print(f"Year {years}: {balance:.2f}€")

print(f"\nIt takes {years} years to double your investment at {rate:.0%} return.")
```

### When to use which loop?

- **`for`** — when you know *what* you're iterating over (a list, a range, etc.).
- **`while`** — when you want to keep going *until* some condition changes.

## 6. Functions

A **function** is a reusable block of code that performs a specific task.  
Think of it like a recipe: you define it once, then "call" it whenever you need it.

You've already used built-in functions like `print()`, `len()`, and `type()`.  
Now let's learn how to write our own.

```python
# Defining a simple function
def greet(name):
    """Prints a personalized greeting."""
    print(f"Hello, {name}! Welcome to the course.")

# Calling the function
greet("Alice")
greet("Bob")
```

```python
# Functions can RETURN a value (instead of just printing)
def calculate_tax(price, tax_rate=0.08):
    """
    Calculate the total price including tax.
    
    Parameters:
        price     - the base price (required)
        tax_rate  - the tax rate as a decimal (default: 0.08 = 8%)
    
    Returns:
        The total price after tax.
    """
    total = price * (1 + tax_rate)
    return total

# Using the function
bill = calculate_tax(100)          # Uses default 8% tax
print(f"With default tax: {bill}€")

bill_875 = calculate_tax(100, 0.0875)  # 8.75% tax rate
print(f"With specified tax rate: {bill_875}€")
```

```python
# Another practical example: ROI calculator
def roi(gain, cost):
    """Calculate Return on Investment as a percentage."""
    return ((gain - cost) / cost) * 100

campaign_roi = roi(gain=75000, cost=50000)
print(f"Campaign ROI: {campaign_roi:.1f}%")
```

### Key takeaways about functions

- `def` starts the function definition, followed by the name and parameters in parentheses.
- The docstring (text in triple quotes) describes what the function does — it is a good habit to write documentation, especially for code that is used by more than one person.
- `return` sends a value back to the caller. Without it, the function returns `None`.
- Parameters can have **default values** (like `tax_rate=0.08`), making specifying a value for them optional when you call the function.

## 7. Objects, Methods, and Classes

You've actually been working with **objects** this entire time.
In Python, *everything* is an object — every string, every number, every list.

An **object** bundles together:
- **Data** (called *attributes*) — the information it holds
- **Behavior** (called *methods*) — the actions it can perform

A **method** is just a function that belongs to a specific object. You call it with dot notation: `object.method()`.

```python
# Strings are objects — they come with many useful methods
company = "KONE Corporation"

print(company.upper())          # "KONE CORPORATION"
print(company.title())          # "Kone Corporation"
print(company.replace("KONE", "MEGA"))   # "MEGA Corporation"
print(company.split())          # ['KONE', 'Corporation']
```

```python
# Lists are objects too
scores = [88, 92, 75, 95, 80]

scores.sort()              # Sorts in place
print(scores)              # [75, 80, 88, 92, 95]
print(scores.count(88))    # How many times does 88 appear?
scores.reverse()           # Reverse the order
print(scores)
```

### Defining your own class

A **class** is a *blueprint* for creating objects.  
Think of it like a template: "Employee" is the class; "Alice" and "Bob" are individual *instances* (objects) of that class.

```python
class Employee:
    """A simple Employee class for our company."""
    
    def __init__(self, name, department, salary):
        """
        Initialize a new Employee.
        __init__ runs automatically when you create an instance.
        'self' refers to the specific object being created.
        """
        self.name = name
        self.department = department
        self.salary = salary
    
    def introduce(self):
        """Print a short introduction."""
        print(f"Hi, I'm {self.name} from {self.department}.")
    
    def annual_bonus(self, bonus_rate=0.10):
        """Calculate the annual bonus."""
        return self.salary * bonus_rate
```

```python
# Creating objects (instances) from the class
emp1 = Employee("Alice", "Marketing", 85000)
emp2 = Employee("Bob", "Finance", 92000)

# Using methods
emp1.introduce()
emp2.introduce()

# Accessing attributes
print(f"\n{emp1.name}'s salary: {emp1.salary:}€")
print(f"{emp1.name}'s bonus:  {emp1.annual_bonus():.0f}€")
print(f"{emp2.name}'s bonus (15%): {emp2.annual_bonus(0.15):.0f}€")
```

### Understanding the key terms

| Term | Meaning | In our example |
|------|---------|----------------|
| **Class** | The blueprint / template | `Employee` |
| **Object / Instance** | A specific thing built from the blueprint | `emp1`, `emp2` |
| **Attribute** | Data stored in the object | `emp1.name`, `emp1.salary` |
| **Method** | A function that belongs to the object | `emp1.introduce()` |
| **`self`** | Refers to "this particular instance" inside the class | Used in every method |
| **`__init__`** | The special method that runs when an object is created | Sets up attributes |
